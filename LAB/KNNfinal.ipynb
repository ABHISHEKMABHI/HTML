```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv("Classified Data",index_col=0)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS',axis=1))
```

```python
scaled_fetures = scaler.transform(df.drop('TARGET CLASS',axis=1))
```

```python
df_feat=pd.DataFrame(scaled_fetures,columns=df.columns[:-1])
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(scaled_fetures,df['TARGET CLASS'],test_size=0.30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred=knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
error_list=[]
for i in range(1,40):
   knn=KNeighborsClassifier(n_neighbors=i)
   knn.fit(X_train,y_train)
   pred_i=knn.predict(X_test)
   error_list.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_list,color="blue",linestyle="dashed",marker='o',markerfacecolor='red',markersize=11)
plt.title="Error rate vs K value"
plt.xlabel=("K VALUE")
plt.ylabel=( "Eror rate")
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
print("with k value=1")
print("\n")
print(confusion_matrix(y_test,pred))
print("\n")
print(classification_report(y_test,pred))
```

```python
knn=KNeighborsClassifier(n_neighbors=23)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
print("with k value=23")
print("\n")
print(confusion_matrix(y_test,pred))
print("\n")
print(classification_report(y_test,pred))
```
